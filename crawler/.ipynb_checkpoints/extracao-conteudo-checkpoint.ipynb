```python
import logging
import urllib3
from bs4 import BeautifulSoup
```

```python
logger = logging.getLogger(__name__)
logging.basicConfig(format='{levelname:.3s}: {message}', style='{', level=logging.INFO)
```

```python
# Desabilitando os Warnings referentes a urls "https"
urllib3.disable_warnings(urllib3.exceptions.InsecureRequestWarning)

http = urllib3.PoolManager()
pagina = http.request('GET', 'https://pt.wikipedia.org/wiki/Linguagem_de_programa%C3%A7%C3%A3o')
```

```python
def getContentFromHtml(html):
    logger.info('Extraindo o conteúdo do html')
    sopa = BeautifulSoup(html, 'lxml')
    for tags in sopa(['script', 'stytle', 'link', 'meta', 'head']):
        tags.decompose()

    conteudo = ' '.join(sopa.stripped_strings)
    logger.info('Extração do conteúdo do html finalizada...')
    return conteudo 
```

```python
sopa = BeautifulSoup(pagina.data, 'lxml')
```

```python
from bs4 import BeautifulSoup
import urllib3
```

```python
# Desabilitando os Warnings referentes a urls "https"
urllib3.disable_warnings(urllib3.exceptions.InsecureRequestWarning)

http = urllib3.PoolManager()
#pagina = http.request('GET', 'https://pt.wikipedia.org/wiki/Linguagem_de_programação')
pagina = http.request('GET', 'https://pt.wikipedia.org/wiki/Linguagem_de_programa%C3%A7%C3%A3o')
pagina.status
```

```python
sopa = BeautifulSoup(pagina.data, 'lxml')
sopa
```

```python
sopa.title
sopa.title.string
```

Trabalhando com `links`

```python
# Listando os links
links = sopa.find_all('a')

# Exibindo a quantidade de links
print('Foram encontrados {} links.'.format(len(links)))

for link in links:
    print(link.get('href'))
    print(link.contents)
```

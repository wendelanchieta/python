```python
from urllib3 import PoolManager
```

Classe `PoolManager` é uma abstração de um pool de conexões para páginas web. <br/>
Isso significa `multiplas requisições` para multiplos `hosts`. 

```python
http = PoolManager()
pagina = http.request('GET', 'http://www.valec.gov.br/dou')
#pagina = http.request('GET', 'http://www.enap.gov.br/')
#pagina = http.request('GET', 'https://www.google.com/', headers={'Accept-Encoding': 'br'})
```

Verificando o `status-code` de retono e exibindo os dados retornados da página

```python
pagina.status
pagina.data
```
